```python
import pandas as pd
import requests
import json
```

```python
cities = pd.read_csv('CleanedSheet.csv')
cities.head()
```

```python
loc = cities[['STATE/UT', 'DISTRICT']]
loc.head()
```

```python
loc['DISTRICT'].nunique()
```

```python
def getCoordinates(location, cache, iteration):
    if location not in cache.keys():
        print("Not found in cache")
        params = {
            'key': 'ciGNJ4Y3fvsCmD3Wrnx8gG7RAWpAWzRg',
            'inFormat': 'kvp',
            'outFormat': 'json',
            'thumbMaps': 'false',
            'location': location
        }
        res = requests.get('http://www.mapquestapi.com/geocoding/v1/address', params=params)
        jsonResponse = res.json()
        if len(jsonResponse['results'][0]['locations']) > 0:
            cache[location] = jsonResponse['results'][0]['locations'][0]['latLng']
    if iteration%10==0:
        print("Iteration {it} successful".format(it=iteration))
    if location in cache.keys():
        return cache[location]
    else:
        return ""
```

```python
cache = {}
district = loc['DISTRICT']
state = loc['STATE/UT']
loc_arr = ["{dist}, {st}".format(dist=district[i], st=state[i]) for i in range(len(district))]
```

```python
for count, elem in enumerate(loc_arr, 0):
    getCoordinates(elem, cache, count)
```

```python
latitude = [ cache[loc]['lat'] if loc in cache.keys() else "" for loc in loc_arr]
assert len(latitude) == len(loc_arr)
```

```python
longitude = [ cache[loc]['lng'] if loc in cache.keys() else "" for loc in loc_arr]
assert len(longitude) == len(loc_arr)
```

```python
cities['latitude'] = latitude
cities['longitude'] = longitude
```
